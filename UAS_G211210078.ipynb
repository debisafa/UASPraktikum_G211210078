```python
#Baca dataset
dataset = pd.read_csv('Stocks.csv')
dataset
```

```python
#a. Distribusi dataset pada setiap atribut Purchase Price, Current Price, dan Gain/Loss dengan grafik histogram
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(dataset['Purchase Price ($)'], bins=10, color="grey")
plt.xlabel('Purchase Price ($)')
plt.ylabel('Frekuensi')
plt.title('Histogram dari Purchase Price ($)')

plt.subplot(1, 3, 2)
plt.hist(dataset['Current Price ($)'], bins=10, color="purple")
plt.xlabel('Current Price ($)')
plt.ylabel('Frekuensi')
plt.title('Histogram dari Current Price ($)')

plt.subplot(1, 3, 3)
plt.hist(dataset['Gain/Loss ($)'], bins=10, color="green")
plt.xlabel('Gain/Loss($)')
plt.ylabel('Frekuensi')
plt.title('Histogram dari Gain/Loss ($)')

plt.tight_layout()
plt.show()
```

```python
# Menyaring Data berdasarkan Dividend / Shares
data = dataset.sort_values(by="Dividend/Shares", ascending=False)
# 5 Data Teratas dari Dividend / Shares
top_5 = data.head(5)
# Plot Data
plt.figure(figsize=(10, 5))
sns.barplot(x="Stock Name", y="Dividend/Shares", data=top_5)
plt.title("5 Data Teratas dari Dividend / Shares")
plt.show()
```

```python
# Menyaring Data berdasarkan Market Value
data = dataset.sort_values(by="Market Value ($)", ascending=False)
# 5 Data Teratas dari Market Value
top_5 = data.head(5)
# Plot Data
plt.figure(figsize=(10, 5))
sns.barplot(x="Stock Name", y="Market Value ($)", data=top_5)
plt.title("5 Data Teratas dari Market Value")
plt.show()
```

```python
# Sort data by Annual Yield
data = dataset.sort_values(by="Annual Yield (%)", ascending=False)
# 5 Data Teratas dari Annual Yield
top_7 = data.head(7)
# Plot Data
plt.figure(figsize=(15, 5))
sns.barplot(x="Stock Name", y="Annual Yield (%)", data=top_7)
plt.title("7 Data Teratas dari Market Value")
plt.show()
```
